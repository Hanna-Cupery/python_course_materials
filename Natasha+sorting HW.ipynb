```python
#1 
def nameRules(text):
    #rules to filter out false-positives for names:
    words = text.split()
    print(words)
    for x in words: 
        if len(x) <2: 
            return False
        if re.search('[А-Я]', x[0]):
            return True 
        else:
            return False 
    
```

```python
#5
names =[] 
adrs=[]
dates=[]
mons =[]



for x in sents:
    prepare_ents =[]
    text = x
    name_ents = (list(ne(text)))
    adr_ents = (list(ae(text)))
    date_ents = (list(de(text)))
    mon_ents = (list(me(text)))
    for ent in name_ents:
        name = text[ent.start:ent.stop]
        if(nameRules(name)):
            names.append((name, 'passed'))
            prepare_ents.append((ent.start, ent.stop, 'Name'))
        else:
            names.append((name, 'failed'))
        
    for ent in adr_ents:
        adr = text[ent.start:ent.stop]
        adrs.append(adr)
        prepare_ents.append((ent.start, ent.stop, 'Address'))
    for ent in date_ents:
        date = text[ent.start:ent.stop]
        dates.append(date)
        prepare_ents.append((ent.start, ent.stop, 'Date'))
    for ent in mon_ents:
        money = text[ent.start:ent.stop]
        mons.append(money)
        prepare_ents.append((ent.start, ent.stop, 'Money'))
        
    show_span_line_markup(
    text, 
    prepare_ents)
        
```

```python
#3
def sort(arr):
    arr =[x.lower() for x in arr]
    for i in range(len(arr)-1):
        for j in range(len(arr)-i-1):
            if(arr[j]>arr[j+1]):
                arr[j+1], arr[j] = arr[j], arr[j+1]
    return arr 
```

```python
#4
adr_following_n =[]
n_following_adr =[]
for x in sents:
    text = x
    name_ents = list(ne(text))
    adr_ents = list(ae(text))
    for n_ent in name_ents:
        if nameRules(text[n_ent.start:n_ent.stop]):
            for a_ent in adr_ents:
                if a_ent.start == (n_ent.stop+1):
                    adr_following_n.append(text[n_ent.start:a_ent.stop])
                if n_ent.start == (a_ent.stop+1):
                    n_following_adr.append(text[a_ent.start:n_ent.stop])

   
```

```python
print('Addresses following names:')
print(adr_following_n)
print('names following addresses:')
print(n_following_adr)
```
